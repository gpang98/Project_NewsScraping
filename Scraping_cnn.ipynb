```python
# To run Chrome
browser = Browser('chrome')
```

```python
url = 'http://www.cnn.com/'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Scrape all quotes on a single page
quotes = soup.find_all('div', class_='layout__content-wrapper layout-homepage__content-wrapper')


for quote in quotes:
    print(quote.text)
```

```python
# Scrape all quotes on a single page
quotes = soup.find_all('div', class_='zone zone--t-light zone-2-observer')


all_quotes_text = ' '.join([quote.text.strip() for quote in quotes])
print(all_quotes_text)
```
